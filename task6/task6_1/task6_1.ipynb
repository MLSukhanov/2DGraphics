```python
import os

def find_files_without_text(directory, text):
    files = os.listdir(directory)

    text_files = [file for file in files if file.endswith('.txt')]

    files_without_text = []

    for file in text_files:
        file_path = os.path.join(directory, file)
        with open(file_path, 'r', encoding='utf-8') as f:
            content = f.read()
            if text not in content:
                files_without_text.append(file)

    return files_without_text

directory = 'dataset/VisDrone2019-DET-val/annotations/'
text_to_find = ',4,'
files_without_text = find_files_without_text(directory, text_to_find)

print(f"Файлы, в которых нет текста '{text_to_find}':")
for file in files_without_text:
    print(file)
```

```python
import numpy as np
import cv2
import os
import random
import torch
import torch.nn as nn
import torchvision.models as models
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix

class EnhancedCarClassifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.resnet = models.resnet18(weights='DEFAULT')
        for param in list(self.resnet.parameters())[:-4]:
            param.requires_grad = False

        self.resnet.fc = nn.Sequential(
            nn.Linear(self.resnet.fc.in_features, 256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, 1)
        )

    def forward(self, x):
        x = self.resnet(x)
        return torch.sigmoid(x)

class CarDetectionDataset(Dataset):
    def __init__(self, images_path, annotations_path, transform=None, balance_classes=True):
        self.images_path = images_path
        self.annotations_path = annotations_path
        self.transform = transform
        self.image_files = []
        self.labels = []
        pos_samples = []
        neg_samples = []

        for f in os.listdir(images_path):
            if f.endswith('.jpg'):
                anno_path = os.path.join(annotations_path, f.replace('.jpg', '.txt'))
                label = 0
                if os.path.exists(anno_path):
                    with open(anno_path, 'r') as anno_file:
                        for line in anno_file:
                            if int(line.split(',')[5]) == 4:  # Class 4 is car in VisDrone
                                label = 1
                                break

                if label == 1:
                    pos_samples.append((f, label))
                else:
                    neg_samples.append((f, label))

        if balance_classes:
            min_samples = min(len(pos_samples), len(neg_samples))
            self.image_files = [x[0] for x in pos_samples[:min_samples]] + [x[0] for x in neg_samples[:min_samples]]
            self.labels = [x[1] for x in pos_samples[:min_samples]] + [x[1] for x in neg_samples[:min_samples]]
        else:
            self.image_files = [x[0] for x in pos_samples + neg_samples]
            self.labels = [x[1] for x in pos_samples + neg_samples]

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        img_path = os.path.join(self.images_path, self.image_files[idx])
        image = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB)
        label = self.labels[idx]

        if self.transform:
            image = self.transform(image)

        return image, torch.tensor([label], dtype=torch.float32)

def train_model():
    train_transform = transforms.Compose([
        transforms.ToPILImage(),
        transforms.Resize((256, 256)),
        transforms.RandomHorizontalFlip(p=0.5),
        transforms.RandomRotation(10),
        transforms.ColorJitter(brightness=0.1, contrast=0.1, saturation=0.1, hue=0.05),
        transforms.RandomAffine(degrees=0, translate=(0.1, 0.1)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    val_transform = transforms.Compose([
        transforms.ToPILImage(),
        transforms.Resize((256, 256)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    full_dataset = CarDetectionDataset(
        "dataset2/BalancedDataset/images",
        "dataset2/BalancedDataset/annotations",
        transform=None,
        balance_classes=True
    )

    all_image_files = full_dataset.image_files.copy()
    all_labels = full_dataset.labels.copy()
    train_files, val_files, train_labels, val_labels = train_test_split(
        all_image_files,
        all_labels,
        test_size=0.3,
        stratify=all_labels,
        random_state=42
    )

    train_dataset = CarDetectionDataset(
        "dataset2/BalancedDataset/images",
        "dataset2/BalancedDataset/annotations",
        transform=train_transform,
        balance_classes=False
    )
    train_dataset.image_files = train_files
    train_dataset.labels = train_labels

    val_dataset = CarDetectionDataset(
        "dataset2/BalancedDataset/images",
        "dataset2/BalancedDataset/annotations",
        transform=val_transform,
        balance_classes=False
    )
    val_dataset.image_files = val_files
    val_dataset.labels = val_labels

    train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True, num_workers=2)
    val_loader = DataLoader(val_dataset, batch_size=4, shuffle=False, num_workers=2)

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = EnhancedCarClassifier().to(device)
    criterion = nn.BCELoss()
    optimizer = torch.optim.AdamW(model.parameters(), lr=1e-3, weight_decay=1e-4, betas=(0.9, 0.999))
    scheduler = torch.optim.lr_scheduler.CosineAnnealingWarmRestarts(optimizer, T_0=10, T_mult=2)

    best_acc = 0
    history = {'train_loss': [], 'val_loss': [], 'accuracy': [],
               'precision': [], 'recall': [], 'f1': []}
    amount_epoch = 20

    for epoch in range(amount_epoch):
        model.train()
        running_loss = 0.0

        for images, labels in tqdm(train_loader, desc=f'Epoch {epoch+1}/{amount_epoch}'):
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()

            running_loss += loss.item() * images.size(0)

        epoch_loss = running_loss / len(train_loader.dataset)
        history['train_loss'].append(epoch_loss)

        model.eval()
        val_loss = 0.0
        all_preds = []
        all_labels = []

        with torch.no_grad():
            for images, labels in val_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)

                val_loss += loss.item() * images.size(0)
                preds = (outputs > 0.5).float()
                all_preds.extend(preds.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())

        val_loss = val_loss / len(val_loader.dataset)
        history['val_loss'].append(val_loss)

        accuracy = np.mean(np.array(all_preds) == np.array(all_labels))
        precision = precision_score(all_labels, all_preds)
        recall = recall_score(all_labels, all_preds)
        f1 = f1_score(all_labels, all_preds)

        history['accuracy'].append(accuracy)
        history['precision'].append(precision)
        history['recall'].append(recall)
        history['f1'].append(f1)

        scheduler.step(accuracy)

        if accuracy > best_acc:
            best_acc = accuracy
            torch.save(model.state_dict(), 'best_car_detector.pth')

        print(f'Epoch {epoch+1}: '
              f'Train Loss: {epoch_loss:.4f}, Val Loss: {val_loss:.4f}, '
              f'Accuracy: {accuracy:.4f}, F1: {f1:.4f}')

    plt.figure(figsize=(15, 5))
    plt.subplot(1, 3, 1)
    plt.plot(history['train_loss'], label='Train Loss')
    plt.plot(history['val_loss'], label='Val Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 3, 2)
    plt.plot(history['accuracy'], label='Accuracy')
    plt.title('Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.subplot(1, 3, 3)
    plt.plot(history['precision'], label='Precision')
    plt.plot(history['recall'], label='Recall')
    plt.plot(history['f1'], label='F1 Score')
    plt.title('Validation Metrics')
    plt.xlabel('Epoch')
    plt.ylabel('Score')
    plt.legend()

    plt.tight_layout()
    plt.savefig('training_metrics.png')
    plt.show()

    cm = confusion_matrix(all_labels, all_preds)
    plt.figure()
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('Confusion Matrix')
    plt.colorbar()
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.xticks([0, 1], ['No Car', 'Car'])
    plt.yticks([0, 1], ['No Car', 'Car'])

    for i in range(2):
        for j in range(2):
            plt.text(j, i, str(cm[i, j]), ha='center', va='center', color='white' if cm[i, j] > cm.max()/2 else 'black')

    plt.savefig('confusion_matrix.png')
    plt.show()

def visualize_detections(model_path, num_samples=5):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = EnhancedCarClassifier().to(device)
    model.load_state_dict(torch.load(model_path, map_location=device))
    model.eval()

    transform = transforms.Compose([
        transforms.ToPILImage(),
        transforms.Resize((256, 256)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    dataset = CarDetectionDataset(
        "dataset2/BalancedDataset/images",
        "dataset2/BalancedDataset/annotations",
        transform=transform,
        balance_classes=True
    )

    indices = random.sample(range(len(dataset)), num_samples)

    plt.figure(figsize=(15, 3*num_samples))
    for i, idx in enumerate(indices):
        image_tensor, label = dataset[idx]
        image = image_tensor.numpy().transpose((1, 2, 0))
        image = image * np.array([0.229, 0.224, 0.225]) + np.array([0.485, 0.456, 0.406])
        image = np.clip(image, 0, 1)

        with torch.no_grad():
            output = model(image_tensor.unsqueeze(0).to(device))
            prob = output.item()
            pred = 1 if prob > 0.5 else 0

        plt.subplot(num_samples, 1, i+1)
        plt.imshow(image)
        plt.title(f'Actual: {"Car" if label.item() == 1 else "No Car"} | '
                  f'Predicted: {"Car" if pred == 1 else "No Car"} ({prob:.2f})')
        plt.axis('off')

    plt.tight_layout()
    plt.savefig('detection_examples.png')
    plt.show()

if __name__ == "__main__":
    if not os.path.exists('best_car_detector.pth'):
        print("Training car detection model...")
        train_model()
    else:
        print("Loading pre-trained model...")

    visualize_detections('best_car_detector.pth', num_samples=7)
```
