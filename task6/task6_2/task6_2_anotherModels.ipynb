```python
import numpy as np
import cv2
import os
import random
import torch
import torch.nn as nn
#from torch.cuda.amp import autocast, GradScaler
from torch import autocast
from torch.amp import GradScaler
import torchvision.models as models
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.metrics import mean_absolute_error, mean_squared_error

class EnhancedCarCounter(nn.Module):
    def __init__(self):
        super().__init__()
        self.backbone = models.efficientnet_v2_m(weights='DEFAULT').features
        for param in list(self.backbone.parameters())[:15]:
            param.requires_grad = False

        self.attention = nn.Sequential(
            nn.Conv2d(1280, 512, 1),
            nn.ReLU(),
            nn.Conv2d(512, 1280, 1),
            nn.Sigmoid()
        )

        self.regressor = nn.Sequential(
            nn.AdaptiveAvgPool2d(1),
            nn.Flatten(),
            nn.Linear(1280, 1024),
            nn.GELU(),
            nn.Dropout(0.5),
            nn.Linear(1024, 512),
            nn.GELU(),
            nn.LayerNorm(512),
            nn.Linear(512, 1)
        )

        self.intermediate_features = nn.ModuleList([
            nn.Sequential(
                nn.Conv2d(160, 256, 3, padding=1),
                nn.ReLU(),
                nn.AdaptiveAvgPool2d(1),
                nn.Flatten()
            ),
            nn.Sequential(
                nn.Conv2d(1280, 512, 3, padding=1),
                nn.ReLU(),
                nn.AdaptiveAvgPool2d(1),
                nn.Flatten()
            )
        ])

        self.final_combiner = nn.Sequential(
            nn.Linear(769, 512),
            nn.GELU()
        )

    def forward(self, x):
        with autocast(device_type="cuda"):
            features = []
            for i, layer in enumerate(self.backbone):
                x = layer(x)
                if i in [4, 8]:
                    features.append(x)

            att = self.attention(x)
            x = x * att

            main_feature = self.regressor(x)

            intermediate_feats = [f(features[i]) for i, f in enumerate(self.intermediate_features)]
            combined = torch.cat([main_feature] + intermediate_feats, dim=1)
            combined = self.final_combiner(combined)

            output = self.regressor[-3:](combined)
            return output

class CarCountingDataset(Dataset):
    def __init__(self, images_path, annotations_path, transform=None):
        self.images_path = images_path
        self.annotations_path = annotations_path
        self.transform = transform
        self.image_files = []
        self.labels = []

        for f in os.listdir(images_path):
            if f.endswith('.jpg'):
                anno_path = os.path.join(annotations_path, f.replace('.jpg', '.txt'))
                count = 0
                if os.path.exists(anno_path):
                    with open(anno_path, 'r') as anno_file:
                        for line in anno_file:
                            parts = line.strip().split(',')
                            if len(parts) >= 6 and int(parts[5]) == 4:
                                count += 1
                self.image_files.append(f)
                self.labels.append(count)

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        img_path = os.path.join(self.images_path, self.image_files[idx])
        image = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB)
        label = self.labels[idx]

        if self.transform:
            image = self.transform(image)

        return image, torch.tensor([label], dtype=torch.float32)

def train_model():
    train_transform = transforms.Compose([
        transforms.ToPILImage(),
        transforms.Resize((640, 640)),
        transforms.RandomHorizontalFlip(p=0.5),
        transforms.RandomAffine(degrees=5, translate=(0.1, 0.1), scale=(0.9, 1.1)),
        transforms.ColorJitter(brightness=0.1, contrast=0.1, saturation=0.1),
        transforms.GaussianBlur(3, sigma=(0.1, 2.0)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    val_transform = transforms.Compose([
        transforms.ToPILImage(),
        transforms.Resize((640, 640)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    train_dataset = CarCountingDataset(
        "dataset2/BalancedDataset/images",
        "dataset2/BalancedDataset/annotations",
        transform=train_transform
    )

    val_dataset = CarCountingDataset(
        "dataset2/BalancedDataset/images",
        "dataset2/BalancedDataset/annotations",
        transform=val_transform
    )

    train_loader = DataLoader(train_dataset, batch_size=2, shuffle=True, num_workers=4, pin_memory=True)
    val_loader = DataLoader(val_dataset, batch_size=2, shuffle=False, num_workers=4)
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = EnhancedCarCounter().to(device)
    criterion = nn.HuberLoss()
    optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-4)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=2)

    history = {'train_loss': [], 'val_loss': [], 'mae': [], 'mse': []}
    best_mae = float('inf')

    scaler = GradScaler()

    for epoch in range(20):
        model.train()
        train_loss = 0.0

        for images, labels in tqdm(train_loader, desc=f'Epoch {epoch+1}/20'):
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()
            with torch.autocast(device_type="cuda", dtype=torch.float16):
                outputs = model(images)
                loss = criterion(outputs, labels)

            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
            train_loss += loss.item() * images.size(0)

        model.eval()
        val_loss = 0.0
        all_preds = []
        all_labels = []

        with torch.no_grad():
            for images, labels in val_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)

                val_loss += loss.item() * images.size(0)
                all_preds.extend(outputs.cpu().numpy().flatten())
                all_labels.extend(labels.cpu().numpy().flatten())

        train_loss = train_loss / len(train_loader.dataset)
        val_loss = val_loss / len(val_loader.dataset)
        mae = mean_absolute_error(all_labels, all_preds)
        mse = mean_squared_error(all_labels, all_preds)

        history['train_loss'].append(train_loss)
        history['val_loss'].append(val_loss)
        history['mae'].append(mae)
        history['mse'].append(mse)

        scheduler.step(val_loss)

        if mae < best_mae:
            best_mae = mae
            torch.save(model.state_dict(), 'best_car_counter.pth')

        print(f'Epoch {epoch+1}: '
              f'Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}, '
              f'MAE: {mae:.4f}, MSE: {mse:.4f}')

    plt.figure(figsize=(15, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history['train_loss'], label='Train Loss')
    plt.plot(history['val_loss'], label='Val Loss')
    plt.title('Loss Curves')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history['mae'], label='MAE')
    plt.plot(history['mse'], label='MSE')
    plt.title('Validation Metrics')
    plt.xlabel('Epoch')
    plt.ylabel('Score')
    plt.legend()
    plt.savefig('training_metrics.png')
    plt.show()

    plt.figure(figsize=(8, 8))
    plt.scatter(all_labels, all_preds, alpha=0.5)
    plt.plot([0, max(all_labels)], [0, max(all_labels)], 'r--')
    plt.xlabel('Actual Count')
    plt.ylabel('Predicted Count')
    plt.title('Prediction vs Ground Truth')
    plt.savefig('predictions_scatter.png')
    plt.show()

def visualize_detections(model_path, num_samples=5):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = EnhancedCarCounter().to(device)
    model.load_state_dict(torch.load(model_path, map_location=device))
    model.eval()

    transform = transforms.Compose([
        transforms.ToPILImage(),
        transforms.Resize((512, 512)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    dataset = CarCountingDataset(
        "dataset2/BalancedDataset/images",
        "dataset2/BalancedDataset/annotations",
        transform=transform,
    )

    indices = random.sample(range(len(dataset)), num_samples)

    plt.figure(figsize=(15, 3*num_samples))
    for i, idx in enumerate(indices):
        image_tensor, label = dataset[idx]
        image = image_tensor.numpy().transpose((1, 2, 0))
        image = image * np.array([0.229, 0.224, 0.225]) + np.array([0.485, 0.456, 0.406])
        image = np.clip(image, 0, 1)

        with torch.no_grad():
            output = model(image_tensor.unsqueeze(0).to(device))
            pred_count = output.item()

        plt.subplot(num_samples, 1, i+1)
        plt.imshow(image)
        plt.title(f'Actual: {int(label.item())} | Predicted: {pred_count:.1f}')
        plt.axis('off')

    plt.tight_layout()
    plt.savefig('detection_examples.png')
    plt.show()

if __name__ == "__main__":
    if not os.path.exists('best_car_counter.pth'):
        print("Training car detection model...")
        train_model()
    else:
        print("Loading pre-trained model...")

    visualize_detections('best_car_counter.pth', num_samples=5)
```

```python
import numpy as np
import cv2
import os
import random
import torch
import torch.nn as nn
import torchvision.models as models
from torch.utils.data import Dataset, DataLoader, Subset
from torchvision import transforms
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import KFold

class EnhancedCarCounter(nn.Module):
    def __init__(self):
        super().__init__()
        self.resnet = models.resnet34(weights='DEFAULT')
        for param in list(self.resnet.parameters())[:-10]:
            param.requires_grad = False
        self.resnet.fc = nn.Sequential(
            nn.Linear(self.resnet.fc.in_features, 512),
            nn.ReLU(),
            nn.Dropout(0.4),
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(256, 1)
        )

    def forward(self, x):
        return self.resnet(x)

class CarCountingDataset(Dataset):
    def __init__(self, images_path, annotations_path, transform=None, file_list=None):
        self.images_path = images_path
        self.annotations_path = annotations_path
        self.transform = transform
        self.image_files = []
        self.labels = []

        selected_files = set(file_list) if file_list is not None else None

        for f in os.listdir(images_path):
            if f.endswith('.jpg'):
                if selected_files is not None and f not in selected_files:
                    continue
                anno_path = os.path.join(annotations_path, f.replace('.jpg', '.txt'))
                count = 0
                if os.path.exists(anno_path):
                    with open(anno_path, 'r') as anno_file:
                        for line in anno_file:
                            parts = line.strip().split(',')
                            if len(parts) >= 6 and int(parts[5]) == 4:
                                count += 1
                self.image_files.append(f)
                self.labels.append(count)

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        img_path = os.path.join(self.images_path, self.image_files[idx])
        image = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB)
        label = self.labels[idx]

        if self.transform:
            image = self.transform(image)

        return image, torch.tensor([label], dtype=torch.float32)

def train_model():
    train_transform = transforms.Compose([
        transforms.ToPILImage(),
        transforms.Resize((512, 512)),
        transforms.RandomHorizontalFlip(),
        transforms.ColorJitter(brightness=0.1, contrast=0.1),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    val_transform = transforms.Compose([
        transforms.ToPILImage(),
        transforms.Resize((512, 512)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    images_path = "dataset2/BalancedDataset/images"
    annotations_path = "dataset2/BalancedDataset/annotations"
    all_files = [f for f in os.listdir(images_path) if f.endswith('.jpg')]

    k_folds = 5
    kf = KFold(n_splits=k_folds, shuffle=True, random_state=42)

    best_mae_overall = float('inf')
    all_history = {'train_loss': [], 'val_loss': [], 'mae': [], 'mse': []}

    for fold, (train_idx, val_idx) in enumerate(kf.split(all_files)):
        print(f'\nFold {fold+1}/{k_folds}')
        train_files = [all_files[i] for i in train_idx]
        val_files = [all_files[i] for i in val_idx]

        train_dataset = CarCountingDataset(images_path, annotations_path, train_transform, train_files)
        val_dataset = CarCountingDataset(images_path, annotations_path, val_transform, val_files)

        train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True, num_workers=2, pin_memory=True)
        val_loader = DataLoader(val_dataset, batch_size=4, shuffle=False, num_workers=2)

        device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        model = EnhancedCarCounter().to(device)
        criterion = nn.HuberLoss()
        optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-4)
        scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=2)

        history = {'train_loss': [], 'val_loss': [], 'mae': [], 'mse': []}
        best_mae = float('inf')

        for epoch in range(30):
            model.train()
            train_loss = 0.0
            for images, labels in tqdm(train_loader, desc=f'Epoch {epoch+1}/40'):
                images, labels = images.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = model(images)
                loss = criterion(outputs, labels)
                loss.backward()
                torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                optimizer.step()
                train_loss += loss.item() * images.size(0)

            model.eval()
            val_loss = 0.0
            all_preds = []
            all_labels = []
            with torch.no_grad():
                for images, labels in val_loader:
                    images, labels = images.to(device), labels.to(device)
                    outputs = model(images)
                    loss = criterion(outputs, labels)
                    val_loss += loss.item() * images.size(0)
                    all_preds.extend(outputs.cpu().numpy().flatten())
                    all_labels.extend(labels.cpu().numpy().flatten())

            train_loss = train_loss / len(train_dataset)
            val_loss = val_loss / len(val_dataset)
            mae = mean_absolute_error(all_labels, all_preds)
            mse = mean_squared_error(all_labels, all_preds)

            history['train_loss'].append(train_loss)
            history['val_loss'].append(val_loss)
            history['mae'].append(mae)
            history['mse'].append(mse)
            scheduler.step(val_loss)

            if mae < best_mae:
                best_mae = mae
                if mae < best_mae_overall:
                    best_mae_overall = mae
                    torch.save(model.state_dict(), 'best_car_counter.pth')

            print(f'Epoch {epoch+1}: Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}, MAE: {mae:.4f}, MSE: {mse:.4f}')

        for key in all_history:
            all_history[key].extend(history[key])

    plt.figure(figsize=(15, 5))
    plt.subplot(1, 2, 1)
    plt.plot(all_history['train_loss'], label='Train Loss')
    plt.plot(all_history['val_loss'], label='Val Loss')
    plt.title('Loss Curves')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(all_history['mae'], label='MAE')
    plt.plot(all_history['mse'], label='MSE')
    plt.title('Validation Metrics')
    plt.xlabel('Epoch')
    plt.ylabel('Score')
    plt.legend()
    plt.savefig('training_metrics.png')
    plt.show()


def visualize_detections(model_path, num_samples=5):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = EnhancedCarCounter().to(device)
    model.load_state_dict(torch.load(model_path, map_location=device))
    model.eval()

    transform = transforms.Compose([
        transforms.ToPILImage(),
        transforms.Resize((512, 512)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    dataset = CarCountingDataset(
        "dataset2/BalancedDataset/images",
        "dataset2/BalancedDataset/annotations",
        transform=transform
    )

    indices = random.sample(range(len(dataset)), num_samples)
    plt.figure(figsize=(15, 3*num_samples))

    for i, idx in enumerate(indices):
        image_tensor, label = dataset[idx]
        image = image_tensor.numpy().transpose((1, 2, 0))
        image = image * np.array([0.229, 0.224, 0.225]) + np.array([0.485, 0.456, 0.406])
        image = np.clip(image, 0, 1)

        with torch.no_grad():
            output = model(image_tensor.unsqueeze(0).to(device))
            pred_count = output.item()

        plt.subplot(num_samples, 1, i+1)
        plt.imshow(image)
        plt.title(f'Actual: {int(label.item())} | Predicted: {pred_count:.1f}')
        plt.axis('off')

    plt.tight_layout()
    plt.savefig('detection_examples.png')
    plt.show()


if __name__ == "__main__":
    if not os.path.exists('best_car_counter.pth'):
        print("Training car counting model with K-Fold...")
        train_model()
    else:
        print("Loading pre-trained model...")

    visualize_detections('best_car_counter.pth', num_samples=5)
```
